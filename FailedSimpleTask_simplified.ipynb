```python
import numpy as np
import h5py, pickle, torch, random
from matplotlib import pyplot as plt
#from vit.models import SingleDecoder
from torch import nn
```

```python
length = 20
n_batch = 64
dim = 128
n_output = 2
input_dim = 4
```

```python
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.backends.cudnn.benchmark = False
torch.backends.cudnn.deterministic = True
```

```python
class transformer_test(nn.Module):
    def __init__(self, input_dim, d_model, d_out, seq_len):
        super().__init__()
        self.cls_token = nn.Parameter(torch.randn(1, 1, dim))
        self.pos_embed = nn.Parameter(torch.randn(1, seq_len + 1, dim))
        self.input_linear = nn.Linear(input_dim, d_model)
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=8, batch_first=True)
        #batch_first True: (batch, seq, feature) False: (seq, batch, feature)
        self.transformer = nn.TransformerEncoder(encoder_layer=self.encoder_layer, num_layers=6)
        self.output_linear = nn.Linear(d_model, d_out)
        #softmax = nn.Softmax(dim=-1)

    def forward(self, x):
        #embedding inputs
        x = self.input_linear(x)
        # add CLS token
        B = x.shape[0]
        cls_tokens = self.cls_token.expand(B, -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        #positional embedding -> temporarily omitted
        x = x + self.pos_embed
        #transformer
        x = self.transformer(x)
        x = self.output_linear(x)
        x = x[:, 0, :]
        return x
        #return softmax(x)
        # output: (n_batch, n_class)
```

```python
#model = SingleDecoder(input_dim=input_dim, length=length, dim=dim, num_classes=n_output+1, mlp_head=True, patch_size=1).cuda()
model = transformer_test(input_dim, dim, n_output+1, length).cuda()
#criterion = MultiCrossEntropyLoss()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
ls = 0
loss_save = []
for i in range(1000):
    optimizer.zero_grad()
    # Inputの0,2番が同期活動（Baseline+2）しているタイミングがあるならTargetの0番が1
    # Inputの1,3番が同期活動（Baseline+2）しているタイミングがあるならTargetの1番が1
    # 両方同期していないならTargetの2が1
    #inputs = torch.rand((n_batch, 1, length, input_dim)).cuda()
    inputs = torch.rand((n_batch, length, input_dim)).cuda()
    # Targetの0番と1番の結果を決める
    t = np.random.randint(2, size=(n_batch, 2)) == 1
    # for k in range(n_batch):
    #     if t[k, 0] and t[k, 1]:
    #         if np.random.rand()<0.5:
    #             t[k, 0]=False
    #         else:
    #             t[k, 1]=False
    # 活動（Baseline+2）させる位置を決める
    r = np.array([[np.random.permutation(length) for _ in range(n_batch)] for _ in range(2)])
    for j in range(2):
        for k in range(n_batch):
            # Inputのj(= 0 or 1)番を2か所活動させる
            inputs[k, r[j, k, :2], j] += 2
            # Targetのj(= 0 or 1)番が1のとき、Inputのj+2(= 2 or 3)番を1か所、Inputのj番と同じところを活動させる
            if t[k, j]:
                inputs[k, r[j, k, 0], j+2] += 2
            # Targetのj(= 0 or 1)番が0のときは、Inputのj+2(= 2 or 3)番の、Inputのj番とは異なるところを1か所活動させる
            else:
                inputs[k, r[j, k, 2], j+2] += 2
            # Inputのj+2(= 2 or 3)番を追加でもう1か所活動させる
            inputs[k, r[j, k, -1], j+2] += 2

    # Targetを数値で表現する（dim=1方向の和が1になる）
    target = np.zeros((n_batch, n_output+1)).astype(np.float32)
    target[t[:, 0], 0] = 1
    target[t[:, 1], 1] = 1
    target[target.sum(axis=1) == 0, 2] = 1
    target /= target.sum(axis=1).reshape(-1, 1)
    target = torch.tensor(target).cuda()
    
    output = model(inputs)
    loss = criterion(output, target)
    loss.backward()
    optimizer.step()
    ls += loss.item()
    if (i + 1) % 100 == 0:
        print("iter: ", i+1, ", loss: ", ls)
        loss_save.append(ls)
        ls = 0
```
